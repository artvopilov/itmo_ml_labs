# Apply linear regression

### Gradient descent

```python
def predict(weights, x_sample):
    return x_sample.dot(weights)


def calculate_smape(p, y):
    return 100 * (abs(p - y) / (abs(p) + abs(y) + 10e-5)).mean()


def calculate_mape(p, y):
    return abs(p - y).mean()


def calculate_mse(p, y):
    return ((p - y) ** 2).mean()


def calculate_gradient_for_smape(weights, x_batch, y_batch):
#     print('Weights')
#     print(weights)
#     print('X batch')
#     print(x_batch)
#     print('weights')
#     print(weights.shape)
    p = predict(weights, x_batch)
#     print('Predictions')
#     print(p)
#     print(p[0])
#     print('Y batch')
#     print(y_batch)
    gradients = 100 * (x_batch * np.sign(p - y_batch) * (abs(p) + abs(y_batch)) 
                       - np.sign(p) * x_batch * abs(p - y_batch)) / ((abs(p) + abs(y_batch)) ** 2 + 10e-5)
    return gradients + 0.001 * weights.reshape(1, -1)


def calculate_gradient_for_mape(weights, x_batch, y_batch):
#     print('Weights')
#     print(weights)
#     print('X batch')
#     print(x_batch)
    p = predict(weights, x_batch)
#     print('Predictions')
#     print(p)
#     print('Y batch')
#     print(y_batch)
    return np.sign(p - y_batch) * x_batch


def calculate_gradient_for_mse(weights, x_batch, y_batch):
    p = predict(weights, x_batch)
    return (p - y_batch) * x_batch


def update_weights(x_batch, y_batch, weights, lr, calculate_grad):
    weights_gradients = calculate_grad(weights, x_batch, y_batch).mean(axis=0).reshape(weights.shape)
#     print('Gradients')
#     print(weights_gradients)
    weights -= lr * weights_gradients
    return weights, weights_gradients


def transform_weights(weights, train_df_offset, train_df_scale):
    x_offset = train_df_offset[:-1]
    x_scale = train_df_scale[:-1]
    y_offset = train_df_offset[-1]
    y_scale = train_df_scale[-1]
    
    transformed_weights = weights[:-1] / x_scale * y_scale
    bias_weight = (weights[-1] - (weights[:-1] * x_offset / x_scale).sum()) * y_scale + y_offset
    transformed_weights = np.append(transformed_weights, bias_weight)
    return transformed_weights
    
```

```python
epochs = 300
batch_size = 16
lr = 1e-2
```

```python
train_errors = []
test_errors = []

for e_i in range(epochs):

    for b_i in range(0, len(train_df_norm), batch_size):
#         print('Weights')
#         print(weights)
        batch = train_df_norm.iloc[b_i:b_i + batch_size]
#         print('batch')
#         print(batch)
        weights, grads = update_weights(batch.drop('target', axis=1).to_numpy(),
                                        batch['target'].to_numpy().reshape(len(batch), 1),
                                        weights,
                                        lr,
                                        calculate_gradient_for_mape)
#         print('weights shape')
#         print(weights.shape)

    p_train = predict(weights, train_df_norm.drop('target', axis=1).to_numpy())
    train_errors.append(calculate_smape(p_train, train_df_norm['target'].to_numpy().reshape(len(train_df_norm), 1)))
        
    transformed_weights = transform_weights(weights, 
                                            train_df_offset.to_numpy().reshape(len(train_df_offset), 1), 
                                            train_df_scale.to_numpy().reshape(len(train_df_offset), 1))
    
    p = predict(transformed_weights, test_df.drop('target', axis=1).to_numpy())
    test_errors.append(calculate_smape(p, test_df['target'].to_numpy().reshape(len(test_df), 1)))
#     print(p[:10])
#     print(test_df['target'][:10])
    
#     print('Epoch ' + str(e_i))
#     print('Grads sum: ' + str(grads.sum()))
#     print('SMAPE: ' + str(calculate_smape(p, test_df['target'].to_numpy().reshape(len(test_df), 1))))
#     print('MAPE: ' + str(calculate_mape(p, test_df['target'].to_numpy().reshape(len(test_df), 1))))
#     print('MSE: ' + str(calculate_mse(p, test_df['target'].to_numpy().reshape(len(test_df), 1))))
    
    if e_i % 50 == 0:
        lr *= 5e-1
#         print(lr)
print('Epoch ' + str(e_i))
print('Grads sum: ' + str(grads.sum()))
print('SMAPE: ' + str(calculate_smape(p, test_df['target'].to_numpy().reshape(len(test_df), 1))))
print('MAPE: ' + str(calculate_mape(p, test_df['target'].to_numpy().reshape(len(test_df), 1))))
print('MSE: ' + str(calculate_mse(p, test_df['target'].to_numpy().reshape(len(test_df), 1))))
```
