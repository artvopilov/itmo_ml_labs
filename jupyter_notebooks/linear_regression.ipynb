```python
dataset_path = dataset_base_path + '/2.txt'
```

```python
with open(dataset_path, 'r') as f:
    factors_num = int(f.readline().strip('\n'))
    
    train_samples_num = int(f.readline().strip('\n'))
    train_samples = []
    for _ in range(train_samples_num):
        sample = list(map(float, f.readline().strip('\n').split()))
        train_samples.append(sample)
        
    test_samples_num = int(f.readline().strip('\n'))
    test_samples = []
    for _ in range(test_samples_num):
        sample = list(map(float, f.readline().strip('\n').split()))
        test_samples.append(sample)
        
```

```python
train_df = pd.DataFrame(train_samples, columns=list(range(factors_num)) + ['target'])
test_df = pd.DataFrame(test_samples, columns=list(range(factors_num)) + ['target'])
```
